```python
import os
import random
import cv2
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
```

```python
TRAIN_DIR = "C:/Users/User/Downloads/CIFAR-10-images-d7d05618061fe37d830fb0419063d113c575ccbc/CIFAR-10-images-d7d05618061fe37d830fb0419063d113c575ccbc/train"
CATEGORIES = []
for c in os.listdir(TRAIN_DIR):
    CATEGORIES.append(c)
print(CATEGORIES)
```

```python
TRAIN_DATA = []
for c in CATEGORIES:
    path = os.path.join(TRAIN_DIR,c)
    class_num = CATEGORIES.index(c)
    for img in tqdm(os.listdir(path)):
        img_arr = cv2.imread(os.path.join(path,img))
        TRAIN_DATA.append([img_arr, class_num])
print(len(TRAIN_DATA))
```

```python
random.shuffle(TRAIN_DATA)
plt.figure(figsize=(20,10))

for i in range(50):
    plt.subplot(5,10,i+1)
    plt.imshow(TRAIN_DATA[i][0])
    plt.xlabel(CATEGORIES[TRAIN_DATA[i][1]])
    plt.xticks([])
    plt.yticks([])
    if i==50:
        break
plt.show()
```

```python
f0 = TRAIN_DATA[0:1000]
f1 = TRAIN_DATA[1000:2000]
f2 = TRAIN_DATA[2000:3000]
f3 = TRAIN_DATA[3000:4000]
f4 = TRAIN_DATA[4000:5000]
```

```python
import math 

midL1=[]
midL2=[]
K=[2,4,6,8,9]

for i in range(5):
    total_acc1 = []
    total_acc2 = []
    
    if i==0:
        train = f1+f2+f3+f4
        validation = f0 
    elif i==1:
        train = f0+f2+f3+f4
        validation = f1
    elif i==2:
        train = f1+f0+f3+f4
        validation = f2
    elif i==3:
        train = f1+f2+f0+f4
        validation = f3
    elif i==4:
        train = f1+f2+f3+f0
        validation = f4
        

    for k in K:
        acc1=0
        acc2=0
        for v in tqdm(validation):
            distances1 = []
            distances2 = []
            
            for t in train:
                distance_L2 = np.sum(np.sqrt(t[0]**2-v[0]**2))
                distances2.append([distance_L2,t[1]])

                distance_L1 = np.sum(t[0]-v[0])
                distances1.append([distance_L1,t[1]])
                
            sorted_distances1 = sorted(distances1, key=lambda a: a[0])
            sorted_distances2 = sorted(distances2, key=lambda a: a[0])

            initial_k1 = [sorted_distances1[i][1] for i in range(k)]
            initial_k2 = [sorted_distances2[i][1] for i in range(k)]
            
            label1 = max(set(initial_k1), key=initial_k1.count)
            label2 = max(set(initial_k2), key=initial_k2.count)

            true_label = v[1]
            if label1 == true_label:
                acc1 += 1
                
            if label2 == true_label:
                acc2 += 1



        accuracy_l1 = acc1 / len(validation) * 100
        accuracy_l2 = acc2 / len(validation) * 100
        
        print(f"Accuracy of L2: {accuracy_l1:.2f}% Accuracy L1: {accuracy_l2:.2f}%")
        total_acc1.append(accuracy_l1)
        total_acc2.append(accuracy_l2)
    
    mid_1 = sum(total_acc1)/5
    mid_2 = sum(total_acc2)/5
    midL1.append(mid_1)
    midL2.append(mid_2)

fig,(ax1,ax2)=plt.subplots(nrows=2,ncols=1)

ax1.plot(K,midL1,label='Accuracy for L1')
ax1.set_xlabel('K values')
ax1.set_ylabel('Accuracy for midpoint')
ax1.legend(loc='best')

ax2.plot(K,midL2,label='Accuracy for L2')
ax2.set_xlabel('K values')
ax2.set_ylabel('Accuracy for midpoint')
ax2.legend(loc='best')

plt.title('Prediction of K values & accuracy')

plt.tight_layout()
plt.show()
```
